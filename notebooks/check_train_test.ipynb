```python
import pandas as pd
```

```python
ratings_give_dir = "../data/csv/ratings_given.csv"
ratings_test_dir = "../data/csv/ratings_test_eval.csv"
```

```python
given_df = pd.read_csv(ratings_give_dir, names=["u_id", "i_id", "rating"])
test_df = pd.read_csv(ratings_test_dir, names=["ID", "u_id", "i_id"])
```

```python
given_df
```

```python
test_df
```

```python
# Find the unique i_id values in each DataFrame
given_u_ids = set(given_df['u_id'])
test_u_ids = set(test_df['u_id'])

# Compute the differences
u_ids_only_in_given = given_u_ids - test_u_ids  # i_id in given_df but not in test_df
u_ids_only_in_test = test_u_ids - given_u_ids  # i_id in test_df but not in given_df

# Print or use the results
print("i_id only in given_df:", u_ids_only_in_given)
print("i_id only in test_df:", u_ids_only_in_test)
```

```python
# Find the unique i_id values in each DataFrame
given_i_ids = set(given_df['i_id'])
test_i_ids = set(test_df['i_id'])

# Compute the differences
i_ids_only_in_given = given_i_ids - test_i_ids  # i_id in given_df but not in test_df
i_ids_only_in_test = test_i_ids - given_i_ids  # i_id in test_df but not in given_df

# Print or use the results
print("i_id only in given_df:", i_ids_only_in_given)
print("i_id only in test_df:", i_ids_only_in_test)
```

```python
known_i_ids_mask = ~test_df['i_id'].isin(i_ids_only_in_test)
known_i_ids_mask
```

```python
test_df_known_i_ids = test_df[known_i_ids_mask]
test_df_known_i_ids
```

```python
test_df_unknown_i_ids = test_df[~known_i_ids_mask]
test_df_unknown_i_ids
```

```python
test_df_known_i_ids.to_csv("../data/csv/ratings_test_eval_known_i_ids.csv", index=False)
test_df_unknown_i_ids.to_csv("../data/csv/ratings_test_eval_unknown_i_ids.csv", index=False)
```
